```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Draw square of random points
x_square = np.random.uniform(0,1,100)
y_square = np.random.uniform(0,1,100)

plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.scatter(x_square, y_square)
```

```python
# draw circle
# circle equation (x-a)^2 + (y-b)^2 = r^2
# where (a,b) is the coordinate of the circle's center and r the circle's radius

x_circle = np.linspace(0, 1, num=50)

# we create a function that take x and give us the y coordinate
def y_funk(x, a=0.5, b=0.5, r=0.5):
    sqrt = lambda x: x**0.5
    term1 = sqrt((r**2-(x-a)**2))
    y1 = -term1 + b
    y2 = term1 + b
    return y1, y2
```

```python
y1_circle, y2_circle = y_funk(x_circle)
```

```python
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.scatter(x_square, y_square, color='k')
plt.plot(x_circle, y1_circle, color='b')
plt.plot(x_circle, y2_circle, color='b')
```

```python
# finally we find the points that are in the circle
# which means the resultant of a point needs to be equal or smaller than the radius
# thanks to pythagore if we have x and y the coordinates of a point we can deduce the resultant which the hypothenus

def c_funk(x, y, r=0.5):
    assert len(x) == len(y), "x and y do not have the same length!"
    
    sqrt = lambda x: x**0.5
    
    x_out = []
    y_out = []
    
    for i in range(len(x)):
        x_recenter = x[i] - 0.5
        y_recenter = y[i] - 0.5
        if sqrt(x_recenter**2 + y_recenter**2) <= r:
            x_out.append(x[i])
            y_out.append(y[i])
    
    return x_out, y_out
```

```python
x_in, y_in = c_funk(x_square, y_square)
```

```python
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.scatter(x_square, y_square, color='k')
plt.plot(x_circle, y1_circle, color='b')
plt.plot(x_circle, y2_circle, color='b')
plt.scatter(x_in, y_in, color='r')
plt.grid(True)
```
