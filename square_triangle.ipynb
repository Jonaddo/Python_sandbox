```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
# draw circle
# circle equation (x-a)^2 + (y-b)^2 = r^2
# where (a,b) is the coordinate of the circle's center and r the circle's radius

x_circle = np.linspace(0, 1, num=50)

# we create a function that take x and give us the y coordinate
def y_funk(x, a=0.5, b=0.5, r=0.5):
    sqrt = lambda x: x**0.5
    term1 = sqrt((r**2-(x-a)**2))
    y1 = -term1 + b
    y2 = term1 + b
    return y1, y2
```

```python
y1_circle, y2_circle = y_funk(x_circle)
```

```python
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.plot(x_circle, y1_circle, color='b')
plt.plot(x_circle, y2_circle, color='b')
```

```python
# draw triangle
# a triangle always has 3 angles/points

x_tri = np.array([0, 0.5, 1, 0])
y_tri = np.array([0, 0.5, 0, 0])

plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.plot(x_tri, y_tri, color='r')
plt.plot(x_circle, y1_circle, color='b')
plt.plot(x_circle, y2_circle, color='b')
```

```python
# Now that we know how to draw a circle and a triangle we want to plot a triangle that touches the circle with his angles
# in order words we want a triangle where his 3 points belong to the circle

s = np.random.binomial(1, 0.5, 1)

rdn_triangle_index = np.random.randint(low=0, high=len(x_circle), size=3)
#rdn_index
x_new_tri = [x_circle[i] for i in rdn_triangle_index] + [x_circle[rdn_triangle_index[0]]]

# since y has 2 solutions, we randomly take a y coordinate in the upper and lower part of the circle
if s == 0:
    y_new_tri = [y1_circle[i] for i in rdn_triangle_index] + [y1_circle[rdn_triangle_index[0]]]
else:
    y_new_tri = [y2_circle[i] for i in rdn_triangle_index] + [y2_circle[rdn_triangle_index[0]]]

plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.plot(x_new_tri, y_new_tri, color='r')
plt.plot(x_circle, y1_circle, color='b')
plt.plot(x_circle, y2_circle, color='b')
```
